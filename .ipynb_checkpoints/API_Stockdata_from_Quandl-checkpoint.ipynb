### 評価指標1 (quandl_code API)

  Quandl API キー : `cjkVD6fMByxGe79XJwFg`

  日経平均株価(N225) : `CHRIS/CME_NK2`<br>
    東証市場第一部の225銘柄の平均値に特殊な修正を加えて算出した指数

  東証株価指数(TOPIX) : `OSE/TOPIXMM2019`<br>
    東証市場第一部に上場している内国普通株式銘柄の全て（約1,800銘柄）を対象とする株価指数
  
  ナスダック総合指数(NASDAQ) : `CHRIS/CME_NQ2`<br>
    アメリカの電子株式市場である”NASDAQ（ナスダック）”に上場している全銘柄を対象とする株価指数<br>
    GAFAMなど、世界的なIT関連企業が上場しているため、これらの業界の動向を読む上での重大指標<br>
    近年では中国企業の上場が伸びており、世界的な動向が反映されやすい

  ダウ平均株価(Dow_Avg) : `BCB/7809`<br>
    ダウ・ジョーンズ社がUS市場に上場している代表的30銘柄の株価の平均値に特殊な修正を加えて算出した指数

  S&P500(SP500) : `MULTPL/SP500_DIV_MONTH`<br>
    スタンダード・アンド・プアーズ社がNYSE、NYSE MKT、NASDAQの銘柄から、代表的500銘柄を選定して算出

  ドル為替レート(USDJPY) : `PERTH/USD_JPY_D`<br>
    日別の円/ドル為替レート



  日銀短観(BOJ_TANKAN) : `BOJ/COAEG2011401CFY40000`<br>
    正式名称「全国企業短期経済観測調査」 統計法に基づいて日本銀行が行う統計調査<br>
    全国の約1万社の企業を対象に、四半期ごとに実施
  
  財務省国債金利情報1年 (JGB_Interest_Rates_1y) : `MOFJ/INTEREST_RATE_JAPAN_1Y`<br>
    債券の価格（金利）と株価は表裏一体の関係<br>
    「債券価格が上昇すると株価は下落する」 逆に「債券価格が下落すると株価は上昇する」

  消費者物価指数(Consumer_Price_Index) : `RATEINF/CPI_JPN`<br>
    総務省統計局が毎月製作<br>
    家計の消費構造を固定し、これに要する費用が物価の変動によってどう変化するかの指数値

  輸出入物価指数(Im_Export_Price_Index) : `BLSN/EIUIH14221`<br>
    生産者物価指数の先行指標、消費者物価指数への波及効果が高いためインフレ動向を探る事が可能な指数<br>
    為替相場変動の影響を大きく受ける特徴がある

  ビッグマック指数(Big_Mac_Index) : `ECONOMIST/BIGMAC_JPN`<br>

    一物一価の法則を用いて購買力平価を求めることができる

  金相場(Gold_Market) : `PERTH/GOLD_JPY_D`<br>
    金融不安が起きたときでも自身の財産を守りやすいという特徴から<br>
    株価が上昇すれば金の相場は下がり、そして株価が低くなれば金の相場は上がる特性がある<br>

  米国財務省実質長期金利(Real_LongTerm_Rates) : `USTREASURY/REALLONGTERM`<br>
    資本の期待実質収益率もしくは実質期待成長率と等しい<br>

### 用語一覧

- Open : 始値
- High : 高値
- Low : 安音
- Last : 終値
- Change : 前日比
- Settle : 清算値(SQ値)
- Sett Price : 清算値
- Volume : 出来高
- Open Interest : 建玉(たてぎょく)
- Previous Day Open Interest : 建玉(たてぎょく)

- Bid High : 売高値
- Ask High : 買高値
- Bid Low : 売安値
- Ask Low : 買安値
- Bid Average : 売平均
- Ask Average : 買平均

- local_price : 現地価格
- dollar_ex　: ドルレート
- dollar_price : ドル価格
- dollar_ppp(purchasing power parity) : 米ドル購買力平価
- dollar_valuation : 米ドル評価額
- dollar_adj_valuation : 米ドル平均評価額
- euro_adj_valuation : 欧州ユーロ平均評価額
- sterling_adj_valuation : UKポンド平均評価額
- yen_adj_valuation : 日本円平均評価額
- yuan_adj_valuation　: 中国元平均評価額

- 6 Month Gold Lease Rates (%PA) : 6ヶ月金リースレート(%)
- LT Real Average (>10Yrs) : 米国実質長期金利(10年以上)

```python
# ライブラリのインポート

import quandl
import itertools
import warnings
import datetime

import numpy as np
import pandas as pd
```

```python
# 各種設定

# .py, .ipynb, .csv保存ディレクトリを設定
Stockdata_path = "/Users/KojiroT/github/ML_Trading_Algorithm/Stock_Data"
Algorithm_path = "/Users/KojiroT/github/ML_Trading_Algorithm"

# QuandlのAPIキー認証
quandl.ApiConfig.api_key = 'cjkVD6fMByxGe79XJwFg'

# 列の表示オプションを100まで表示可能に変更
pd.options.display.max_columns = 100

# 取得したい日付レンジを指定
start = datetime.datetime(2017, 1, 1)
end = datetime.datetime(2019, 12, 31)

# ユーティリティ関数

def DF_EXCHANGER(raw_df):
    # 日付設定：2017 - 2019
    df = (raw_df['2017':'2019'])
    # change(不要なもの)がカラムに入っていた場合は削除
    if "Change" in df.columns:
        df = df.drop("Change", axis=1)
    else:
        pass
    # カラム設定
    JP_columns = {'Date':'日付','Open':'始値', 'High':'高値','Low':'安値','Last':'終値','Settle':'清算値','Sett Price':'清算値',
                  'Volume':'出来高','Previous Day Open Interest':'建玉','Open Interest':'建玉','Value':'指数','RATE':'レート',
                  'Bid High':'売高値','Ask High':'買高値','Bid Low':'売安値','Ask Low':'買安値','Bid Average':'売平均','Ask Average':'買平均',
                  'local_price':'現地価格','dollar_ex':'ドルレート','dollar_price':'米ドル価格','dollar_ppp':'米ドル購買力平価','dollar_valuation':'米ドル評価額',
                  'dollar_adj_valuation':'米ドル平均評価額','euro_adj_valuation':'欧州ユーロ平均評価額','sterling_adj_valuation':'UKポンド平均評価額',
                  'yen_adj_valuation':'日本円平均評価額','yuan_adj_valuation':'中国元平均評価額',
                  '6 Month Gold Lease Rates (%PA)':'6ヶ月金リースレート(%)','LT Real Average (>10Yrs)':'米国実質長期金利(10年以上)'}
    # 日本語名に変更
    df = df.rename(columns=JP_columns)
    # 前日比・翌日比算出のために終値を抜き出す(終値が入っていない場合はpassする)
    if "終値" in df.columns:
        change_base = df['終値']
        change_base = pd.DataFrame(change_base)
        # pct_change()で変化率を算出する
        Beforeday_Change = change_base.pct_change()
        Beforeday_Change = Beforeday_Change.rename(columns={'終値': '前日比'})
        # pct_change(-1)で1行後の列の差分の変化率を算出する
        Nextday_Change = change_base.pct_change(-1)
        Nextday_Change = Nextday_Change.rename(columns={'終値': '翌日比'})
        # データフレームに前日比と翌日比を結合させる
        df = pd.concat([df, Beforeday_Change, Nextday_Change], axis=1)
    else:
        pass
    # 欠損値処理(欠損値を前の値で置き換えた後、後の値で置き換える)
    df = df.fillna(method='ffill').fillna(method='bfill')
    return df
```
